```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Download nltk resources if not downloaded
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')


```

```python
# Load dataset
dataset = pd.read_csv("/content/Update_Background_set.CSV")

# Text preprocessing function
def preprocess_text(text):
    # Lowercase the text
    text = text.lower()

    # Remove punctuation and special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Tokenize the text
    tokens = word_tokenize(text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    # Join tokens back into text
    preprocessed_text = ' '.join(tokens)

    return preprocessed_text

# Apply preprocessing to the 'Scripts' column
dataset['Preprocessed_Scripts'] = dataset['Scripts'].apply(preprocess_text)

# Save preprocessed dataset to a new CSV file
dataset.to_csv("/content/preprocessed_dataset.csv", index=False)

print("Text preprocessing completed and saved to preprocessed_dataset.csv")
```

```python

import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize
from nltk.sentiment import SentimentIntensityAnalyzer

# Download NLTK resources
nltk.download('punkt')
nltk.download('vader_lexicon')

# Load the dataset
data = pd.read_csv('/content/preprocessed_dataset.csv')

# Assuming 'Script' column contains the script data
script_texts = data['Scripts']

# Initialize sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Function to perform sentiment analysis on each script
def analyze_sentiment(script_text):
    sentiments = []
    for script in script_text:
        # Tokenize script into sentences
        sentences = sent_tokenize(script)
        script_sentiment = 0
        # Calculate sentiment score for each sentence and aggregate
        for sentence in sentences:
            sentiment_score = sia.polarity_scores(sentence)
            script_sentiment += sentiment_score['compound']
        # Average sentiment score for the entire script
        script_sentiment /= len(sentences) if len(sentences) > 0 else 1
        sentiments.append(script_sentiment)
    return sentiments

# Perform sentiment analysis on script texts
script_sentiments = analyze_sentiment(script_texts)

# Add sentiment scores to DataFrame
data['Sentiment'] = script_sentiments

# Save the updated DataFrame to a new CSV file or any other desired output
data.to_csv('sentiment_analysis_results.csv', index=False)
```

```python
import pandas as pd

# Load the dataset with sentiment analysis results
sentiment_dataset = pd.read_csv("/content/sentiment_analysis_results.csv")

# Function to classify sentiment based on scores
def classify_sentiment(score):
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Ensure 'Sentiment' column contains numeric values
sentiment_dataset['Sentiment'] = pd.to_numeric(sentiment_dataset['Sentiment'], errors='coerce')

# Drop rows with NaN values in 'Sentiment' column
sentiment_dataset.dropna(subset=['Sentiment'], inplace=True)

# Apply sentiment classification to the 'Sentiment' column
sentiment_dataset['Sentiment_Class'] = sentiment_dataset['Sentiment'].apply(classify_sentiment)

# Save the modified dataset back to a CSV file
sentiment_dataset.to_csv("/content/sentiment_analysis_results_with_class.csv", index=False)

```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
import spacy

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

# Load English language model for spaCy
nlp = spacy.load("en_core_web_sm")

# Load the CSV dataset
dataset = pd.read_csv("/content/sentiment_analysis_results_with_class.csv")

# Function to extract named entities (character names) from the script text
def extract_entities(script):
    doc = nlp(script)
    entities = [ent.text for ent in doc.ents if ent.label_ == 'PERSON']
    # Remove duplicates
    entities = list(set(entities))
    return entities

# Function to extract actions, emotions, and tone from the script text
def extract_actions_emotions_tone(script):
    # Tokenize the script
    tokens = word_tokenize(script)
    # Perform part-of-speech tagging
    tagged_tokens = nltk.pos_tag(tokens)
    # Extract verbs (actions) and adjectives (emotions/tone)
    actions = [word for word, pos in tagged_tokens if pos.startswith('VB')]
    emotions_tone = [word for word, pos in tagged_tokens if pos.startswith('JJ')]
    return actions, emotions_tone

# Apply the extraction functions to the 'Scripts' column and store the results in new columns
dataset['Characters'] = dataset['Scripts'].apply(extract_entities)
dataset['Actions'], dataset['Emotions_Tone'] = zip(*dataset['Scripts'].apply(extract_actions_emotions_tone))

# Optionally, you can save the modified dataset back to a CSV file
dataset.to_csv("extracted_info_dataset.csv", index=False)

# Print the first few rows of the dataset with extracted information
print(dataset.head())
```

```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Assuming you have already installed transformers and pandas

# Initialize the tokenizer and model with an appropriate model
tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")

def paraphrase(text, num_beams=5, num_return_sequences=1,
               repetition_penalty=1.0, no_repeat_ngram_size=2,
               temperature=1.0, max_length=64):
    input_ids = tokenizer.encode(text, return_tensors="pt", max_length=max_length, truncation=True)

    paraphrases = model.generate(input_ids, max_length=max_length, num_beams=num_beams,
                                 temperature=temperature, repetition_penalty=repetition_penalty,
                                 num_return_sequences=num_return_sequences, no_repeat_ngram_size=no_repeat_ngram_size)

    paraphrased_text = tokenizer.decode(paraphrases[0], skip_special_tokens=True)
    return paraphrased_text

# Load your dataset
df = pd.read_csv('/content/preprocessed_dataset.csv')

# Apply the paraphrasing function to each row in the 'Scripts' column
# Assuming your column name is 'Scripts', not 'Preprocessed_Scripts'
df['Paraphrased_Scripts'] = df['Scripts'].apply(lambda x: paraphrase(x))

# Save the modified DataFrame to a new CSV file
df.to_csv('/content/para_file.csv', index=False)
```
